```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df = pd.read_csv('/content/retail_store_inventory.csv')
```

```python
df.head()
```

```python
df = df.drop(['Product ID','Store ID'],axis = 1)
```

```python
df = df.drop('Date',axis = 1)
df.head()
```

```python
# @title Units Sold

from matplotlib import pyplot as plt
df['Units Sold'].plot(kind='line', figsize=(8, 4), title='Units Sold')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
le = LabelEncoder()
```

```python
df.head()
```

```python
# prompt: label encode category region and seasonality columns

# Create LabelEncoder objects
le_region = LabelEncoder()
le_seasonality = LabelEncoder()

# Fit and transform the 'Region' column
df['Region'] = le_region.fit_transform(df['Region'])

# Fit and transform the 'Seasonality' column
df['Seasonality'] = le_seasonality.fit_transform(df['Seasonality'])

df.head()
```

```python
df['Category'] = le_region.fit_transform(df['Category'])
df['Weather Condition'] = le_region.fit_transform(df['Weather Condition'])
df.head()
```

```python
y = df['Demand Forecast']
x = df.drop('Demand Forecast',axis = 1)
corr = x.corrwith(y)
corr
```

```python
# prompt: convert corr to all positives and arrange them in descending order

corr_pos = abs(corr)  # Convert all correlations to positive values
corr_sorted = corr_pos.sort_values(ascending=False) # Sort in descending order
corr_sorted
```

```python
fin_feat = corr_sorted.index[:7]
fin_feat
```

```python
x = df[fin_feat]
y = df['Demand Forecast']
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
```

```python
kf = KFold(n_splits=5,shuffle=True,random_state = 42)
```

```python
model = LinearRegression()
```

```python
x_train_ss = StandardScaler().fit_transform(x_train)
x_test_ss = StandardScaler().fit_transform(x_test)
```

```python
model.fit(x_train_ss,y_train)
```

```python
y_pred = model.predict(x_test_ss)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
